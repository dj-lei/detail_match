## 一.汽车之家在售车处理

```python
def delete_useless_char(df):
    text = df['value']
    text = text.replace('&nbsp','')
    text = text.replace('/','')
    text = text.replace(';','')
    return text

f = open("../tmp/train/new.car.info.json", encoding='utf-8')
count = 0
final = pd.DataFrame()
for text_json in f.readlines():
    detail = json.loads(text_json)
    config = detail.pop('config')
    param = detail.pop('param')
    specitems = detail.pop('specitems')
    if len(specitems) == 0:
        continue
    result = pd.DataFrame(specitems)
    result['brand'] = detail['brand']
    result['brand_id'] = detail['brand_id']
    result['type'] = detail['type']
    result['type_id'] = detail['type_id']
    result['manufacturer'] = detail['manufacturer']
    result = result.rename(columns={'id':'specid'})
    count = count + 1
    print(count,'combine',detail['type'])
    for i in range(0, len(param)):
        param_temp = pd.DataFrame(param[i])
        for j in range(0, len(param_temp)):
            paramitems = param_temp.loc[j,'paramitems']
            temp = pd.DataFrame(paramitems['valueitems'])
            temp = temp.rename(columns={'value':paramitems['name']})
            result = result.merge(temp, how='left', on=['specid'])

    for i in range(0, len(config)):
        config_temp = pd.DataFrame(config[i])
        for j in range(0, len(config_temp)):
            configitems = config_temp.loc[j,'configitems']
            temp = pd.DataFrame(configitems['valueitems'])
            temp['value'] = temp.apply(delete_useless_char,axis=1)
            temp = temp.rename(columns={'value':configitems['name']})
            result = result.merge(temp, how='left', on=['specid'])

    final = final.append(result,sort=False)
    final.reset_index(inplace=True, drop=True)
final.to_csv('../tmp/train/car_autohome_new.csv', index=False)
```

## 三.生成汽车之家标准库

```python
def cal_online_year(df):
    return int(re.findall('(\d+)款', df['detail_name'])[0])

def process_price_bn(df):
    if (df['price_bn'] == '暂无报价')|(str(df['price_bn']) == 'nan'):
        return np.NAN
    temp = re.findall(r'万',df['price_bn'])
    if len(temp) > 1:
        text = df['price_bn'].split('~')[1]
    else:
        text = df['price_bn']
    text = text.replace('万','')
    return float(text)

def process_control(df):
    auto = ['自动','双离合','无级','CVT','手自','AMT','电动','-']
    manual = ['手动','序列','ISR',]
    for name in auto:
        if re.findall(name,df['简称']):
            return '自动'
    for name in manual:
        if re.findall(name,df['简称']):
            return '手动'
    return np.NAN

def process_volume(df):
    if (df['energy'] == '纯电动')|(df['energy'] == '增程式'):
        return 0
    text = re.findall(r'(\d\.\d)',df['发动机'])
    if text:
        return text[0]
    return np.NAN

def process_emission_standard(df):
    if (df['energy'] == '纯电动')|(df['energy'] == '增程式'):
        return pd.Series(['E','国VI'])
    return pd.Series([df['volume_extend'],df['emission_standard']])

car_autohome_new = pd.read_csv('../tmp/train/car_autohome_new.csv', low_memory=False)
car_autohome_new = car_autohome_new.loc[:, ['specid', 'brand','brand_id','type','type_id','name','manufacturer','厂商指导价(元)','级别','能源类型','发动机','简称','气缸排列形式','环保标准']]
car_autohome_new = car_autohome_new.rename(columns={'specid':'detail_id','name':'detail_name','brand':'brand_name','type':'model_name'})

car_autohome_stop = pd.read_csv('../tmp/train/car_autohome_stop.csv', low_memory=False)
car_autohome_stop = car_autohome_stop.loc[:, ['id','brand','brand_id','type','type_id','stop_sell_name','manufacturer','厂商指导价(元)','级别','能源类型','发动机','简称','气缸排列形式','环保标准']]
car_autohome_stop = car_autohome_stop.rename(columns={'id':'detail_id','stop_sell_name':'detail_name','brand':'brand_name','type':'model_name'})

car_autohome_all = car_autohome_new.append(car_autohome_stop, sort=False)
car_autohome_all = car_autohome_all.rename(columns={'detail_id':'detail_slug','type_id':'model_slug','brand_id':'brand_slug',
                                                   '厂商指导价(元)':'price_bn','级别':'body','能源类型':'energy','气缸排列形式':'volume_extend','环保标准':'emission_standard'})
car_autohome_all = car_autohome_all.drop_duplicates(['detail_slug']).reset_index(drop=True)
car_autohome_all['online_year'] = car_autohome_all.apply(cal_online_year, axis=1)
car_autohome_all['price_bn'] = car_autohome_all.apply(process_price_bn, axis=1)
car_autohome_all = car_autohome_all.loc[(car_autohome_all['price_bn'].notnull())&(car_autohome_all['price_bn'] > 0), :]
car_autohome_all = car_autohome_all.sort_values(by=['brand_slug','model_slug','detail_slug']).reset_index(drop=True)
car_autohome_all[['volume_extend','emission_standard']] = car_autohome_all.apply(process_emission_standard, axis=1)
car_autohome_all['volume_extend'] = car_autohome_all['volume_extend'].fillna('L')
car_autohome_all['emission_standard'] = car_autohome_all['emission_standard'].fillna('国V')
car_autohome_all['volume'] = car_autohome_all.apply(process_volume, axis=1)
car_autohome_all['control'] = car_autohome_all.apply(process_control, axis=1)
car_autohome_all = car_autohome_all.drop(['发动机','简称'], axis=1)
car_autohome_all.to_csv('../tmp/train/car_autohome_all.csv', index=False)
car_autohome_all.to_csv('/home/ml/ProgramProject/evaluation-predict/tmp/train/car_autohome_all.csv', index=False)
car_autohome_all.info()
```
