## 一.汽车之家在售车处理

```python
def delete_useless_char(df):
    text = df['value']
    text = text.replace('&nbsp','')
    text = text.replace('/','')
    text = text.replace(';','')
    return text

f = open("../tmp/train/car.info.json", encoding='utf-8')
count = 0
final = pd.DataFrame()
for text_json in f.readlines():
    detail = json.loads(text_json)
    config = detail.pop('config')
    param = detail.pop('param')
    specitems = detail.pop('specitems')
    if len(specitems) == 0:
        continue
    result = pd.DataFrame(specitems)
    result['brand'] = detail['brand']
    result['brand_id'] = detail['brand_id']
    result['type'] = detail['type']
    result['type_id'] = detail['type_id']
    result['manufacturer'] = detail['manufacturer']
    result = result.rename(columns={'id':'specid'})
    count = count + 1
    print(count,'combine',detail['type'])
    for i in range(0, len(param)):
        param_temp = pd.DataFrame(param[i])
        for j in range(0, len(param_temp)):
            paramitems = param_temp.loc[j,'paramitems']
            temp = pd.DataFrame(paramitems['valueitems'])
            temp = temp.rename(columns={'value':paramitems['name']})
            result = result.merge(temp, how='left', on=['specid'])

    for i in range(0, len(config)):
        config_temp = pd.DataFrame(config[i])
        for j in range(0, len(config_temp)):
            configitems = config_temp.loc[j,'configitems']
            temp = pd.DataFrame(configitems['valueitems'])
            temp['value'] = temp.apply(delete_useless_char,axis=1)
            temp = temp.rename(columns={'value':configitems['name']})
            result = result.merge(temp, how='left', on=['specid'])

    final = final.append(result,sort=False)
    final.reset_index(inplace=True, drop=True)
final.to_csv('../tmp/train/car_autohome_new.csv', index=False)
```

## 二.汽车之家停售车处理

```python
def delete_useless_char(df):
    text = df['value']
    text = text.replace('&nbsp','')
    text = text.replace('/','')
    text = text.replace(';','')
    return text

f = open("../tmp/train/stop.sell.info.json", encoding='utf-8')
count = 0
final = pd.DataFrame()
for text_json in f.readlines():
    detail = json.loads(text_json)

    result = pd.DataFrame([''],columns=['brand'])
    result['brand'] = detail['brand']
    result['brand_id'] = detail['brand_id']
    result['type'] = detail['type']
    result['type_id'] = detail['type_id']
    result['manufacturer'] = detail['manufacturer']
    result['stop_sell_name'] = detail['stop_sell_name']
    result['sell_type'] = detail['sell_type']
    result['id'] = detail['specid']
    config = detail.pop('result')
    configitems = config['configitems']
    paramitems = config['paramitems']
            
    param = pd.DataFrame(dict(zip(re.findall('(?<=\'name\': \').*?(?=\')',str(paramitems)), re.findall('(?<=\'value\': \').*?(?=\')',str(paramitems)))),index=[0])
    param['id'] = detail['specid']
    result = result.merge(param,how='left',on=['id'])

    config = pd.DataFrame([re.findall('(?<=\'value\': \').*?(?=\')',str(configitems))],columns=re.findall('(?<=\'name\': \').*?(?=\')',str(configitems)))
    config['id'] = detail['specid']
    result = result.merge(config,how='left',on=['id'])
            
    count = count + 1
    print(count,'combine',detail['type'])

    final = final.append(result,sort=False).reset_index(drop=True)

final.to_csv('../tmp/train/car_autohome_stop.csv', index=False)
```

## 三.生成汽车之家标准库

```python
def cal_online_year(df):
    return int(re.findall('(\d+)款', df['detail_name'])[0])

def process_price_bn(df):
    if (df['price_bn'] == '暂无报价')|(str(df['price_bn']) == 'nan'):
        return np.NAN
    temp = re.findall(r'万',df['price_bn'])
    if len(temp) > 1:
        text = df['price_bn'].split('~')[1]
    else:
        text = df['price_bn']
    text = text.replace('万','')
    return float(text)

def process_control(df):
    auto = ['自动','双离合','无级','CVT','手自','AMT','电动','-']
    manual = ['手动','序列','ISR',]
    for name in auto:
        if re.findall(name,df['简称']):
            return '自动'
    for name in manual:
        if re.findall(name,df['简称']):
            return '手动'
    return np.NAN

def process_volume(df):
    if (df['energy'] == '纯电动')|(df['energy'] == '增程式'):
        return 0
    text = re.findall(r'(\d\.\d)',df['发动机'])
    if text:
        return text[0]
    return np.NAN

def process_emission_standard(df):
    if (df['energy'] == '纯电动')|(df['energy'] == '增程式'):
        return pd.Series(['E','国VI'])
    return pd.Series([df['volume_extend'],df['emission_standard']])

def process_listed_year(df):
    if (str(df['listed_year']) == 'nan') | (str(df['listed_year']) == '-'):
        return df['online_year'] - 1
    return str(df['listed_year']).split('.')[0]

car_autohome_new = pd.read_csv('../tmp/train/car_autohome_new.csv', low_memory=False)
car_autohome_new = car_autohome_new.loc[:, ['specid', 'brand','brand_id','type','type_id','name','manufacturer','厂商指导价(元)','级别','能源类型','发动机','简称','气缸排列形式','环保标准','上市时间']]
car_autohome_new = car_autohome_new.rename(columns={'specid':'detail_id','name':'detail_name','brand':'brand_name','type':'model_name'})

car_autohome_stop = pd.read_csv('../tmp/train/car_autohome_stop.csv', low_memory=False)
car_autohome_stop = car_autohome_stop.loc[:, ['id','brand','brand_id','type','type_id','stop_sell_name','manufacturer','厂商指导价(元)','级别','能源类型','发动机','简称','气缸排列形式','环保标准','上市时间']]
car_autohome_stop = car_autohome_stop.rename(columns={'id':'detail_id','stop_sell_name':'detail_name','brand':'brand_name','type':'model_name'})

car_autohome_all = car_autohome_new.append(car_autohome_stop, sort=False)
car_autohome_all = car_autohome_all.rename(columns={'detail_id':'detail_slug','type_id':'model_slug','brand_id':'brand_slug',
                                                   '厂商指导价(元)':'price_bn','级别':'body','能源类型':'energy','气缸排列形式':'volume_extend','环保标准':'emission_standard','上市时间':'listed_year'})
car_autohome_all = car_autohome_all.drop_duplicates(['detail_slug']).reset_index(drop=True)
car_autohome_all['online_year'] = car_autohome_all.apply(cal_online_year, axis=1)
car_autohome_all['price_bn'] = car_autohome_all.apply(process_price_bn, axis=1)
car_autohome_all = car_autohome_all.loc[(car_autohome_all['price_bn'].notnull())&(car_autohome_all['price_bn'] > 0), :]
car_autohome_all = car_autohome_all.sort_values(by=['brand_slug','model_slug','detail_slug']).reset_index(drop=True)
car_autohome_all[['volume_extend','emission_standard']] = car_autohome_all.apply(process_emission_standard, axis=1)
car_autohome_all['volume_extend'] = car_autohome_all['volume_extend'].fillna('L')
car_autohome_all['emission_standard'] = car_autohome_all['emission_standard'].fillna('国V')
car_autohome_all['volume'] = car_autohome_all.apply(process_volume, axis=1)
car_autohome_all['control'] = car_autohome_all.apply(process_control, axis=1)
car_autohome_all['listed_year'] = car_autohome_all.apply(process_listed_year, axis=1)
car_autohome_all['listed_year'] = car_autohome_all['listed_year'].astype(int)
car_autohome_all = car_autohome_all.drop(['发动机','简称'], axis=1)
car_autohome_all.to_csv('../tmp/train/car_autohome_all.csv', index=False)
car_autohome_all.to_csv('/home/ml/ProgramProject/evaluation-predict/tmp/train/car_autohome_all.csv', index=False)
car_autohome_all.info()
```

## 五.生成标准库

```python
from xpinyin import Pinyin

def process_brand_name(df,cur_brand):
    brand_name = cur_brand.loc[(cur_brand['brand_slug'] == df['car_autohome_brand_id']),'brand_name']
    if len(brand_name) == 0:
        brand_name = df['name']
    else:
        brand_name = brand_name.values[0]
    if str(df['slug']) == 'nan':
        return pd.Series([brand_name, 'car_autohome_brand_' + str(df['car_autohome_brand_id'])])   
    return pd.Series([brand_name,df['slug']])

def process_model_name(df,cur_model,new_combine_brand):
    model_name = cur_model.loc[(cur_model['model_slug'] == df['car_autohome_model_id']),'model_name']
    if len(model_name) == 0:
        model_name = df['name']
    else:
        model_name = model_name.values[0]
    brand_slug = cur_model.loc[(cur_model['model_slug'] == df['car_autohome_model_id']),'brand_slug']
    if len(brand_slug) == 0:
        parent = df['parent']
    else:
        parent = new_combine_brand.loc[(new_combine_brand['car_autohome_brand_id'] == brand_slug.values[0]),'slug'].values[0]
    if str(df['slug']) == 'nan':
        return pd.Series([model_name,parent, 'car_autohome_model_' + str(df['car_autohome_model_id'])])   
    return pd.Series([model_name,parent, df['slug']])

def process_detail_name(df,cur_detail,new_combine_model):
    detail_name = cur_detail.loc[(cur_detail['detail_slug'] == df['car_autohome_detail_id']),'detail_name']
    if len(detail_name) == 0:
        detail_name = df['detail_model']
    else:
        detail_name = detail_name.values[0]
    model_slug = cur_detail.loc[(cur_detail['detail_slug'] == df['car_autohome_detail_id']),'model_slug']
    if len(model_slug) == 0:
        global_slug = df['global_slug']
    else:
        global_slug = new_combine_model.loc[(new_combine_model['car_autohome_model_id'] == model_slug.values[0]),'slug'].values[0]
    if str(df['detail_model_slug']) == 'nan':
        return pd.Series([detail_name,global_slug, 'car_autohome_detail_' + str(df['car_autohome_detail_id'])])   
    return pd.Series([detail_name,global_slug, df['detail_model_slug']])

def process_keywords(df):
    p = Pinyin()
    keywords = p.get_pinyin(df['name'], '')
    final_keywords = ','+keywords+','+df['name']+','+keywords
    first_letter = p.get_initial(df['name'][0])
    return pd.Series([final_keywords,first_letter,keywords])

car_autohome_all = pd.read_csv('../tmp/train/car_autohome_all.csv')
cur_brand = car_autohome_all.loc[:,['brand_name','brand_slug']]
cur_brand = cur_brand.drop_duplicates(['brand_slug']).reset_index(drop=True)
cur_model = car_autohome_all.loc[:,['model_name','model_slug','brand_slug','body','manufacturer']].rename(columns={'body':'classified','manufacturer':'mum'})
cur_model = cur_model.drop_duplicates(['model_slug']).reset_index(drop=True)
cur_detail = car_autohome_all.loc[:,['detail_slug','detail_name','model_slug']]
origin_combine_brand = pd.read_csv('../tmp/train/combine_brand.csv')
origin_combine_model = pd.read_csv('../tmp/train/combine_model.csv')
origin_combine_detail = pd.read_csv('../tmp/train/combine_detail.csv')

# 品牌处理
new_brand = cur_brand.loc[~(cur_brand['brand_slug'].isin(list(set(origin_combine_brand.car_autohome_brand_id.values)))),:].rename(columns={'brand_name':'name','brand_slug':'car_autohome_brand_id'})
print('新增品牌数量:',len(new_brand),list(new_brand.name.values))
delete_brand = origin_combine_brand.loc[~(origin_combine_brand['car_autohome_brand_id'].isin(cur_brand.brand_slug.values)),:]
print('删除品牌数量:',len(delete_brand),list(delete_brand.name.values))
new_brand[['keywords', 'first_letter', 'pinyin']] = new_brand.apply(process_keywords, axis=1)
new_brand['brand_area'] = '国产'
new_brand['has_detailmodel'] = 1
new_brand['units'] = 0
new_brand['score'] = 0
new_brand['status'] = 'Y'
new_brand['popular'] = 'B'
new_brand['on_sale'] = '1'
origin_combine_brand.loc[(origin_combine_brand['car_autohome_brand_id'].isin(list(delete_brand.car_autohome_brand_id.values))),'status']='D'
new_combine_brand = origin_combine_brand.append(new_brand,sort=False)
new_combine_brand[['name','slug']] = new_combine_brand.apply(process_brand_name,args=(cur_brand,),axis=1)
new_combine_brand.to_csv('./new_combine_brand.csv', index=False)

# 车系处理
new_model = cur_model.loc[~(cur_model['model_slug'].isin(list(set(origin_combine_model.car_autohome_model_id.values)))),:].rename(columns={'model_name':'name','model_slug':'car_autohome_model_id'})
new_model = new_model.drop(['brand_slug'],axis=1)
print('新增车系数量:',len(new_model),list(new_model.name.values))
delete_model = origin_combine_model.loc[~(origin_combine_model['car_autohome_model_id'].isin(cur_model.model_slug.values)),:]
print('删除车系数量:',len(delete_model),list(delete_model.name.values))
new_model[['keywords', 'first_letter', 'pinyin']] = new_model.apply(process_keywords, axis=1)
new_model['attribute'] = '国产'
new_model['normalized_name'] = new_model['name']
new_model['has_detailmodel'] = 1
new_model['units'] = 0
new_model['score'] = 0
new_model['status'] = 'Y'
new_model['popular'] = 'B'
new_model['on_sale'] = '1'
origin_combine_model.loc[(origin_combine_model['car_autohome_model_id'].isin(list(delete_model.car_autohome_model_id.values))),'status']='D'
new_combine_model = origin_combine_model.append(new_model,sort=False)
new_combine_model[['name','parent','slug']] = new_combine_model.apply(process_model_name,args=(cur_model,new_combine_brand,),axis=1)
new_combine_model.to_csv('./new_combine_model.csv', index=False)
```

```python
def process_emission_standard(df):
    if (str(df['emission_standard']) == 'nan') or (str(df['emission_standard']) == '-'):
        return np.NAN
    text = df['emission_standard'].lower()
    if len(re.findall('vii',text)) > 0:
        return df['emission_standard'][0]+'7'
    elif len(re.findall('vi',text)) > 0:
        return df['emission_standard'][0]+'6'
    elif len(re.findall('iv',text)) > 0:
        return df['emission_standard'][0]+'4'
    elif len(re.findall('v',text)) > 0:
        return df['emission_standard'][0]+'5'
    elif len(re.findall('iii',text)) > 0:
        return df['emission_standard'][0]+'3'
    elif len(re.findall('ii',text)) > 0:
        return df['emission_standard'][0]+'2'
    elif len(re.findall('i',text)) > 0:
        return df['emission_standard'][0]+'1'
    return df['emission_standard'][0]+df['emission_standard'][1]

# 款型处理
new_detail = cur_detail.loc[~(cur_detail['detail_slug'].isin(list(set(origin_combine_detail.car_autohome_detail_id.values)))),:].rename(columns={'detail_slug':'car_autohome_detail_id'})
new_detail = new_detail.drop(['model_slug'],axis=1)
print('新增款型数量:',len(new_detail))
delete_detail = origin_combine_detail.loc[~(origin_combine_detail['car_autohome_detail_id'].isin(cur_detail.detail_slug.values)),:]
print('删除车系数量:',len(delete_detail),list(delete_detail.detail_model.values))

temp = car_autohome_all.loc[:,['detail_name','detail_slug','price_bn','online_year','listed_year','volume','control','emission_standard','volume_extend']].rename(columns={'detail_slug':'car_autohome_detail_id','detail_name':'detail_model','model_slug':'global_slug','online_year':'year'})
new_detail = new_detail.drop(['detail_name'],axis=1)
new_detail = new_detail.merge(temp,how='left',on=['car_autohome_detail_id'])
new_detail['created_on'] = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
new_detail['status'] = 'Y'
new_detail['has_param'] = 'Y'
origin_combine_detail.loc[(origin_combine_detail['car_autohome_detail_id'].isin(list(delete_detail.car_autohome_detail_id.values))),'status']='D'
new_combine_detail = origin_combine_detail.append(new_detail,sort=False).reset_index(drop=True)
new_combine_detail[['detail_model','global_slug','detail_model_slug']] = new_combine_detail.apply(process_detail_name,args=(cur_detail,new_combine_model,),axis=1)
new_combine_detail['emission_standard'] = new_combine_detail['emission_standard'].fillna('国V')
new_combine_detail['emission_standard'] = new_combine_detail.apply(process_emission_standard,axis=1)
new_combine_detail.to_csv('./new_combine_detail.csv',index=False)
```
