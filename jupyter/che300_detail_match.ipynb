```python
# coding=utf-8
import os
import re
import json
import math
import random
import datetime
import pandas as pd
import numpy as np
from sqlalchemy import create_engine

import sys
sys.path.append('../')

from match.match.match import Match
```

```python
result = pd.DataFrame()
car_source = pd.read_csv('./detail_che300.csv', low_memory=False)
car_source['model_name'] = car_source['brand_name'] + ' ' + car_source['model_name']

detail_name = car_source.loc[:, ['model_name']].sort_values(by=['model_name'])
detail_name = detail_name.drop_duplicates(['model_name']).reset_index(drop=True)

print('等待匹配车源数据量:', len(car_source), '款型数量:', len(detail_name))
match = Match()
for i, detail in enumerate(list(detail_name.loc[:, 'model_name'].values)):

    part1 = car_source.loc[(car_source['model_name'] == detail), :].reset_index(drop=True)
    print(i, detail)
    gpj_detail = match.predict(part1['model_name'][0], cos_similar=0)['data']
    if len(gpj_detail) == 0:
        part1['origin_name'] = part1['model_name']
        part1 = part1.drop(['model_name'], axis=1)
        result = result.append(part1, sort=False).reset_index(drop=True)
    else:
        part2 = pd.DataFrame(gpj_detail, index=[0])
        part1['origin_name'] = part1['model_name']
        part1 = part1.drop(['model_name'], axis=1)
        temp = pd.concat([part1, pd.DataFrame(columns=list(part2.columns))], sort=False)
        temp.loc[:, part2.columns] = part2.loc[0, :].values
        result = result.append(temp, sort=False).reset_index(drop=True)

result.to_csv('./che300_match_result.csv', index=False)
miss_match = result.loc[(result['brand_slug'].isnull()),:].reset_index(drop=True)
miss_match.to_csv('./miss_match.csv', index=False)
```
